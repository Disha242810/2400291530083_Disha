```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derv(x):
    return sigmoid(x) * (1-sigmoid(x))
```

```python
X = np.array([[3,9,2,11,0]])
y = np.array([[6,18,4,22,0]])
```

```python
def forward_propagation(X, w1, b1, w2, b2):
    z1 = X*w1 + b1
    a1 = sigmoid(z1)
    z2 = a1*w2 + b2
    return sigmoid(z2)
```

```python
np.random.seed(1)
input_size = 1
hidden_size = 1
output_size = 1
learning_rate = 0.1

w1 = np.random.uniform(size=(input_size, hidden_size))
b1 = np.random.uniform(size=(1,hidden_size))
w2 = np.random.uniform(size=(hidden_size, output_size))
b2 = np.random.uniform(size=(1, output_size))

for epoch in range(5):
    z1 = np.dot(w1, X) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(w2, a1) + b2
    y_pred = sigmoid(z2)

    error = y - y_pred
    derv_y_pred = error * sigmoid_derv(y_pred)

    error_hidden = w2.T.dot(derv_y_pred)
    derv_hidden = error_hidden * sigmoid_derv(z2)

    w2 += a1.T.dot(derv_y_pred) * learning_rate
    b2 += np.sum(derv_y_pred)
    w1 += X.T.dot(derv_hidden) * learning_rate
    b1 += np.sum(derv_hidden, axis=0, keepdims=True)

y_pred.round(3)
```
